## meta > reg > topic 
- 지역별로 나눈뒤 topic별로 100개씩 데이터 추출 작업 
- 100개 데이터는 작은 거 같아서 1,000개로 진행하려고 했으나 지역별로 확인해본 결과 데이터의 개수가 1,000 미만인 데이터들이 존재해서 100개로 변경

```python
# jj
df_jj = pd.DataFrame(columns = ['text','reg','topic'])

for idx, row in df.iterrows():
    row = df.loc[df['reg'] == 'jj' ]
    break
#     for i in row : 
#         i = row.loc[row['topic'] == '4차산업혁명']    
#         break
    
df_jj = pd.concat([df_jj,row])
df_jj.drop_duplicates(inplace=True)
df_jj.sort_values('topic', ascending=True)
```

```python
# kw
df_kw = pd.DataFrame(columns = ['text','reg','topic'])

for idx, row in df.iterrows():
    row = df.loc[df['reg'] == 'kw' ]
    break
    
df_kw = pd.concat([df_kw,row])
df_kw.drop_duplicates(inplace=True)
df_kw.sort_values('topic', ascending=True)
```

```python
# jd
df_jd = pd.DataFrame(columns = ['text','reg','topic'])

for idx, row in df.iterrows():
    row = df.loc[df['reg'] == 'jd' ]
    break
    
df_jd = pd.concat([df_jd,row])
df_jd.drop_duplicates(inplace=True)
df_jd.sort_values('topic', ascending=True)
```

```python
# gs
df_gs = pd.DataFrame(columns = ['text','reg','topic'])

for idx, row in df.iterrows():
    row = df.loc[df['reg'] == 'gs' ]
    break
    
df_gs = pd.concat([df_gs,row])
df_gs.drop_duplicates(inplace=True)
df_gs.sort_values('topic', ascending=True)
```
