## train set 추출

```python
train = pd.merge(raw, test, how = 'outer', indicator = True)
```

```python
train
```

```python
print(train['_merge'].unique())
```

```python
print(train['_merge'].value_counts())
```

```python
train.query('_merge == "left_only"').drop(columns=['_merge'])
```

```python
Train = train.query('_merge == "left_only"').drop(columns=['_merge'])
```

```python
Train
```

```python
Train.to_csv(directory + 'Train_set_sampling.csv' , index=False)
```
